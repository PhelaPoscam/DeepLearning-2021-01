```python
from Utilities.io import DataLoader
from Utilities.painter import Visualizer
from Models.RRDBNet import RRDBNet
```

### Load in the sample images

```python
import glob
DATA_PATH = 'Sam'
loader = DataLoader()
data = loader.load(glob.glob(DATA_PATH + '/*.png'), batchSize=1)
painter = Visualizer()
for downSample, original in data.take(6):
    painter.plot(downSample, original)
```

### Load in the pretrained super-resolution model

```python
# pretrained rrdb network can be found in the Pretrained folder
MODEL_PATH = 'Pretrained/rrdb'
model = RRDBNet(blockNum=10)
model.load_weights('/content/drive/MyDrive/License-super-resolution-master2/Weights/')
```

### Run plate enhancement

```python
for downSample, original in data.take(6):
    yPred = model.predict(original) #change to downSample if you want the downsized image
    painter.plot(original, yPred)
```
